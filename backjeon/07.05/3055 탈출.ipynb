```python
import copy
def check(map_y, map_x, now, y, x):
    if now[0] + y >= 0 and now[0] + y < map_y and now[1] + x >= 0 and now[1] + x < map_x:
        return True
    return False

y, x = map(int, input().split())
forest = []
water = []
for i in range(y):
    maps = list(input())
    if 'D' in maps:
        house = [i, maps.index('D')]
    if 'S' in maps:
        ddg = [i, maps.index('S')]
    if '*' in maps:
        for j in range(x):
            if maps[j] == '*':
                water.append([i, j])
    forest.append(maps)
    
ddg_q = [ddg]
water_q = water

ddg_q_visit = [[False for z in range(x)] for c in range(y)]
ddg_q_visit[ddg[0]][ddg[1]] = True
water_q_visit = [[False for z in range(x)] for c in range(y)]
for i in range(len(water)):
    water_q_visit[water[i][0]][water[i][1]] = True

y_go = [0, 0, 1, -1]
x_go = [1, -1, 0, 0]


count = 1
chk=True
while(ddg_q and chk):
    new_water = []
    for i in range(len(water_q)):
        cur_water = water_q[i]
        now_y = cur_water[0]
        now_x = cur_water[1]
       
        for j in range(4):
            yy = y_go[j]
            xx = x_go[j]
            if check(y, x, [now_y, now_x], yy, xx): #범위 안나감
                if (not water_q_visit[now_y+yy][now_x+xx]) and forest[now_y+yy][now_x+xx] != 'X' and forest[now_y+yy][now_x+xx] != 'D': #안가봤고 벽 아닐 때
                    water_q_visit[now_y+yy][now_x+xx] = True #가준다
                    new_water.append([now_y+yy, now_x+xx]) #추가해줌
    water_q = copy.deepcopy(new_water)
    new_ddg = []
    for i in range(len(ddg_q)):
        cur_ddg = ddg_q[i]
        now_y = cur_ddg[0]
        now_x = cur_ddg[1]
        for j in range(4):
            yy = y_go[j]
            xx = x_go[j]
            if check(y, x, [now_y, now_x], yy, xx):

                if (not ddg_q_visit[now_y+yy][now_x+xx]) and forest[now_y+yy][now_x+xx] != 'X' and (not water_q_visit[now_y+yy][now_x+xx]):
                    #안가봤고, 벽 아니고, 물 아닐 때
                    if forest[now_y+yy][now_x+xx] == 'D':
                        print(count)
                        chk=False

                    ddg_q_visit[now_y+yy][now_x+xx] = True
                    new_ddg.append([now_y+yy, now_x+xx])
    ddg_q = copy.deepcopy(new_ddg)
    count += 1
if(chk):
    print("KAKTUS")
#     print(ddg_q)
#     for i in range(len(water_q_visit)):
#         print(water_q_visit[i])
    

```
