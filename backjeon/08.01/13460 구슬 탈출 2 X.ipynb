```python
import copy
y, x = map(int, input().split())

world = []
for i in range(y):
    l = list(input())
    if 'R' in l:
        R = [i, l.index('R')]
        l[l.index('R')] = '.'
    if 'B' in l:
        B = [i, l.index('B')]
        l[l.index('B')] = '.'
    if 'O' in l:
        O = [i, l.index('O')]
    world.append(l)

    
#어차피 10번 이상 하면 -1 -> 4의 10승 => 1024*1024 = 백만.

step = 0
q = []
q.append([R, B, -1])
r_out = False
b_out = False
while(step < 10 and not r_out and not b_out):
    new_q = []
    end = False
    for ball in q:
        RL = copy.deepcopy(ball[0])
        BL = copy.deepcopy(ball[1])
        r_idx = 0
        b_idx = 0
        left = False
        #왼쪽 1
        if (world[RL[0]][RL[1]-1] in ['.', 'O'] or world[BL[0]][BL[1]-1] in ['.', 'O']) and ball[2]!=1: #둘중 하난 갈수있을때
            left = True
            if RL[0] == BL[0]: #같은 줄에 있을 때
                if RL[1] < BL[1]: #빨간 공이 왼쪽 -> 빨간 공 부터 움직임
                    while world[RL[0]][RL[1]-1-r_idx] == '.':
                        r_idx += 1
                    if world[RL[0]][RL[1]-1-r_idx] == 'O':
                        r_out = True
                        RL[1] -= r_idx + 1
                    elif world[RL[0]][RL[1]-1-r_idx] == '#':
                        RL[1] -= r_idx
                    
                    while world[BL[0]][BL[1]-1-b_idx] == '.':
                        if BL[1] - b_idx - 1 == RL[1]: #갈곳이 빨간공
                            break
                        b_idx += 1
                    if world[BL[0]][BL[1]-1-b_idx] == 'O':
                        b_out = True
                        BL[1] -= b_idx + 1
                    else:
                        BL[1] -= b_idx  
                elif BL[1] < RL[1]: #파란공이 왼쪽
                    while world[BL[0]][BL[1]-1-b_idx] == '.':
                        b_idx += 1
                    if world[BL[0]][BL[1]-1-b_idx] == 'O':
                        b_out = True
                        BL[1] -= b_idx + 1
                    elif world[BL[0]][BL[1]-1-b_idx] == '#':
                        BL[1] -= b_idx
                    
                    while world[RL[0]][RL[1]-1-r_idx] == '.':
                        if RL[1] - r_idx - 1 == BL[1]: #갈곳이 파란공
                            break
                        r_idx += 1
                    if world[RL[0]][RL[1]-1-r_idx] == 'O':
                        r_out = True
                        RL[1] -= r_idx + 1
                    else:
                        RL[1] -= r_idx  
                        
            else: #신경X
                while world[RL[0]][RL[1]-1-r_idx] == '.':
                    r_idx += 1
                if world[RL[0]][RL[1]-1-r_idx] == 'O':
                    r_out = True
                    RL[1] -= r_idx + 1
                elif world[RL[0]][RL[1]-1-r_idx] == '#':
                    RL[1] -= r_idx
                
                while world[BL[0]][BL[1]-1-b_idx] == '.':
                    b_idx += 1
                if world[BL[0]][BL[1]-1-b_idx] == 'O':
                    b_out = True
                    BL[1] -= b_idx + 1
                elif world[BL[0]][BL[1]-1-b_idx] == '#':
                    BL[1] -= b_idx
        temp1 = [RL, BL, 1, r_out, b_out]
        r_out=False
        b_out=False
        #print(temp1)
        RR = copy.deepcopy(ball[0])
        BR = copy.deepcopy(ball[1])
        r_idx = 0
        b_idx = 0
        right = False
        #오른쪽 2
        if (world[RR[0]][RR[1]+1] in ['.', 'O'] or world[BR[0]][BR[1]+1] in ['.', 'O']) and ball[2] != 2: #둘중 하난 갈수있을때
            right = True
            if RR[0] == BR[0]: #같은 줄에 있을 때
                if RR[1] > BR[1]: #빨간 공이 오른쪽 -> 빨간 공 부터 움직임
                    while world[RR[0]][RR[1]+1+r_idx] == '.':
                        r_idx += 1
                    if world[RR[0]][RR[1]+1+r_idx] == 'O':
                        r_out = True
                        RR[1] += r_idx + 1
                    elif world[RR[0]][RR[1]+1+r_idx] == '#':
                        RR[1] += r_idx
                    
                    while world[BR[0]][BR[1]+1+b_idx] == '.':
                        if BR[1] + b_idx + 1 == RR[1]: #갈곳이 빨간공
                            break
                        b_idx += 1
                    if world[BR[0]][BR[1]+1+b_idx] == 'O':
                        b_out = True
                        BR[1] += b_idx + 1
                    else:
                        BR[1] += b_idx  
                elif BR[1] > RR[1]: #파란공이 오른쪽
                    while world[BR[0]][BR[1]+1+b_idx] == '.':
                        b_idx += 1
                    if world[BR[0]][BR[1]+1+b_idx] == 'O':
                        b_out = True
                        BR[1] += b_idx + 1
                    elif world[BR[0]][BR[1]+1+b_idx] == '#':
                        BR[1] += b_idx
                    
                    while world[RR[0]][RR[1]+1+r_idx] == '.':
                        if RR[1] + r_idx + 1 == BR[1]: #갈곳이 파란공
                            break
                        r_idx += 1
                    if world[RR[0]][RR[1]+1+r_idx] == 'O':
                        r_out = True
                        RR[1] += r_idx + 1
                    else:
                        RR[1] += r_idx  
                        
            else: #신경X
                while world[RR[0]][RR[1]+1+r_idx] == '.':
                    r_idx += 1
                if world[RR[0]][RR[1]+1+r_idx] == 'O':
                    r_out = True
                    RR[1] += r_idx + 1
                elif world[RR[0]][RR[1]+1+r_idx] == '#':
                    RR[1] += r_idx
                
                while world[BR[0]][BR[1]+1+b_idx] == '.':
                    b_idx += 1
                if world[BR[0]][BR[1]+1+b_idx] == 'O':
                    b_out = True
                    BR[1] += b_idx + 1
                elif world[BR[0]][BR[1]+1+b_idx] == '#':
                    BR[1] += b_idx
        temp2 = [RR, BR, 2, r_out, b_out]
        r_out=False
        b_out=False
        RU = copy.deepcopy(ball[0])
        BU = copy.deepcopy(ball[1])
        r_idx = 0
        b_idx = 0
        up = False
        #위쪽 3
        if (world[RU[0]-1][RU[1]] in ['.', 'O'] or world[BU[0]-1][BU[1]] in ['.', 'O']) and ball[2] != 3: #둘중 하난 갈수있을때
            up = True
            if RU[1] == BU[1]: #같은 줄에 있을 때
                if RU[0] < BU[0]: #빨간 공이 위쪽 -> 빨간 공 부터 움직임
                    while world[RU[0]-1-r_idx][RU[1]] == '.':
                        r_idx += 1
                    if world[RU[0]-1-r_idx][RU[1]] == 'O':
                        r_out = True
                        RU[0] -= r_idx + 1
                    elif world[RU[0]-1-r_idx][RU[1]] == '#':
                        RU[0] -= r_idx
                    
                    while world[BU[0]-1-b_idx][BU[1]] == '.':
                        if BU[0] - b_idx - 1 == RU[0]: #갈곳이 빨간공
                            break
                        b_idx += 1
                    if world[BU[0]-1-b_idx][BU[1]] == 'O':
                        b_out = True
                        BU[0] -= b_idx + 1
                    else:
                        BU[0] -= b_idx  
                elif B[0] < R[0]: #파란공이 위쪽
                    while world[BU[0]-1-b_idx][BU[1]] == '.':
                        b_idx += 1
                    if world[BU[0]-1-b_idx][BU[1]] == 'O':
                        b_out = True
                        BU[0] -= b_idx + 1
                    elif world[BU[0]-1-b_idx][BU[1]] == '#':
                        BU[0] -= b_idx
                    
                    while world[RU[0]-1-r_idx][RU[1]] == '.':
                        if RU[0] - r_idx - 1 == BU[0]: #갈곳이 파란공
                            break
                        r_idx += 1
                    if world[RU[0]-1-r_idx][RU[1]] == 'O':
                        r_out = True
                        RU[0] -= r_idx + 1
                    else:
                        RU[0] -= r_idx  
                        
            else: #신경X
                while world[RU[0]-1-r_idx][RU[1]] == '.':
                    r_idx += 1
                if world[RU[0]-1-r_idx][RU[1]] == 'O':
                    r_out = True
                    RU[0] -= r_idx + 1
                elif world[RU[0]-1-r_idx][RU[1]] == '#':
                    RU[0] -= r_idx
                
                while world[BU[0]-1-b_idx][BU[1]] == '.':
                    b_idx += 1
                if world[BU[0]-1-b_idx][BU[1]] == 'O':
                    b_out = True
                    BU[0] -= b_idx + 1
                elif world[BU[0]-1-b_idx][BU[1]] == '#':
                    BU[0] -= b_idx
        temp3 = [RU, BU, 3, r_out, b_out]
        r_out=False
        b_out=False
        RD = copy.deepcopy(ball[0])
        BD = copy.deepcopy(ball[1])
        r_idx = 0
        b_idx = 0
        down = False
        #아래쪽 4
        if (world[RD[0]+1][RD[1]] in ['.', 'O'] or world[BD[0]+1][BD[1]] in ['.', 'O']) and ball[2] != 4: #둘중 하난 갈수있을때
            down = True
            if RD[1] == BD[1]: #같은 줄에 있을 때
                if RD[0] > BD[0]: #빨간 공이 아래 -> 빨간 공 부터 움직임
                    while world[RD[0]+1+r_idx][RD[1]] == '.':
                        r_idx += 1
                    if world[RD[0]+1+r_idx][RD[1]] == 'O':
                        r_out = True
                        RD[0] += r_idx + 1
                    elif world[RD[0]+1+r_idx][RD[1]] == '#':
                        RD[0] += r_idx
                    
                    while world[BD[0]+1+b_idx][BD[1]] == '.':
                        if BD[0] + b_idx + 1 == RD[0]: #갈곳이 빨간공
                            break
                        b_idx += 1
                    if world[BD[0]+1+b_idx][BD[1]] == 'O':
                        b_out = True
                        BD[0] += b_idx + 1
                    else:
                        BD[0] += b_idx
                elif BD[0] > RD[0]: #파란공이 아래쪽
                    while world[BD[0]+1+b_idx][BD[1]] == '.':
                        b_idx += 1
                    if world[BD[0]+1+b_idx][BD[1]] == 'O':
                        b_out = True
                        BD[0] += b_idx + 1
                    elif world[BD[0]+1+b_idx][BD[1]] == '#':
                        BD[0] += b_idx
                    
                    while world[RD[0]+1+r_idx][RD[1]] == '.':
                        if RD[0] + r_idx + 1 == BD[0]: #갈곳이 파란공
                            break
                        r_idx += 1
                    if world[RD[0]+1+r_idx][RD[1]] == 'O':
                        r_out = True
                        RD[0] += r_idx + 1
                    else:
                        RD[0] += r_idx  
                        
            else: #신경X
                while world[RD[0]+1+r_idx][RD[1]] == '.':
                    r_idx += 1
                if world[RD[0]+1+r_idx][RD[1]] == 'O':
                    r_out = True
                    RD[0] += r_idx + 1
                elif world[RD[0]+1+r_idx][RD[1]] == '#':
                    RD[0] += r_idx
                
                while world[BD[0]+1+b_idx][BD[1]] == '.':
                    b_idx += 1
                if world[BD[0]+1+b_idx][BD[1]] == 'O':
                    b_out = True
                    BD[0] += b_idx + 1
                elif world[BD[0]+1+b_idx][BD[1]] == '#':
                    BD[0] += b_idx
        temp4 = [RD, BD, 4, r_out, b_out]
        r_out=False
        b_out=False
        #print("L", RL, BL)
        #print("R", RR, BR)
        #print("U", RU, BU)
        #print("D", RD, BD)
        #print(temp1, temp2, temp3, temp4)
        if left and not temp1[4]:
            new_q.append(temp1)
        if right and not temp2[4]:
            new_q.append(temp2)
        if up and not temp3[4]:
            new_q.append(temp3)
        if down and not temp4[4]:
            new_q.append(temp4)
            
        if (temp1[3] and not temp1[4]) or (temp2[3] and not temp2[4]) or (temp3[3] and not temp3[4]) or (temp4[3] and not temp4[4]):
            end = True
            break #빨간공만 나간 케이스 있을 때
    q = copy.deepcopy(new_q)
    #print(q, step)
    step += 1
    if end:
        break
        
if end: #빨간 공만 탈출
    print(step)
else:
    print(-1)
```
