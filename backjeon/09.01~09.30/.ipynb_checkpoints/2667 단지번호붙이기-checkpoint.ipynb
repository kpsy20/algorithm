```python
import copy
n = int(input())
world = []
visited = [[False for x in range(n)] for y in range(n)]
for i in range(n):
    l = list(map(int, input()))
    world.append(l)
    for j in range(n):
        if l[j] == 0:
            visited[i][j] = True
gogo = True
start = []
res = []
x = [0, 0, -1, 1]
y = [1, -1, 0, 0]
while gogo:
    ans = 0
    gogo = False
    for i in range(n):
        for j in range(n):
            if visited[i][j] == False:
                start = [[i, j]]
                gogo = True
                break
        if gogo:
            break
    if not start:
        break
    visited[start[0][0]][start[0][1]] = True
    #print(start)
    while start:
        #print(start)
        #for i in visited:
           #print(i)
        new_start = []
        while start:
            now = start.pop()
            ans += 1
            for i in range(4):
                if now[0]+y[i]>=0 and now[0]+y[i]<n and now[1]+x[i]>=0 and now[1]+x[i]<n and visited[now[0]+y[i]][now[1]+x[i]] == False:
                    visited[now[0]+y[i]][now[1]+x[i]] = True
                    new_start.append([now[0]+y[i], now[1]+x[i]])
        start = copy.deepcopy(new_start)
    res.append(ans)
print(len(res), ' '.join([str(x) for x in sorted(res)]))
        
    
```
