```python
def put(n, last, idx, res): #last = 이전 체스말들 idx 지금 내 col
    cant = []
    for i in range(len(last)):
        cant.append(last[i][0])
        if idx - last[i][1] + last[i][0] < n:
            cant.append(idx-last[i][1]+last[i][0]) 
        
        if last[i][0] - (idx - last[i][1]) >= 0:
            cant.append(last[i][0]-(idx-last[i][1]))
        cant = list(set(cant))
        
    go = [x for x in range(n)]
    result = 0
    if idx == n-1:
        for i in go:
            if i not in cant:
                result += 1
        res.append(result)
        return 0
    print(cant)
    if(len(cant) == n):
        return 0
    for i in go:
        if i not in cant:
            print(last+[[i, idx]])
            put(n, last + [[i, idx]], idx+1, res)
    
    
n = int(input())
ans = 0
for i in range(n):
    result  = []
    print(i, "time")
    put(n, [[i, 0]], 1, result)
    print("result:", result)
    ans += sum(result)
print(ans)
```
